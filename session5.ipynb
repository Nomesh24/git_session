```python
a=pd.read_csv('Visadataset.csv')
a
```

```python
cat=a.select_dtypes(include='object').columns
cat
```

```python
for i in cat[1:]:
    key=a[i].value_counts().keys()
    value=a[i].value_counts().values
    cmap=plt.get_cmap('viridis')
colors1=cmap(np.linspace(0,1,len(key)))
    plt.figure(figsize=(10,8))
    plt.bar(key,value)
    
```

```python
plt.figure(figsize=(28,14))
for i in range(1,9):
    key=a[cat[i]].value_counts().keys()
    value=a[cat[i]].value_counts().values
    plt.subplot(3,3,i)
    plt.bar(key,value)
    
```

```python
plt.figure(figsize=(28,12))
for i in range(1,9):
    key=a[cat[i]].value_counts().keys()
    value=a[cat[i]].value_counts().values
    plt.subplot(3,3,i).bar(key,value)
```

```python
plt.figure(figsize=(28,14))
for i in range(1,9):
    key=a[cat[i]].value_counts().keys()
    value=a[cat[i]].value_counts().values
    plt.subplot(3,3,i).bar(key,value)
    
```

**pie chart**

```python
a['continent'].value_counts(normalize=True)
```

```python
key=a['continent'].value_counts().keys()
value=a['continent'].value_counts().values
```

```python
plt.pie(value,
        labels=key,
       autopct="%0.2f%%")# yha % % btata h ki percentage h aur 0.2f% btata h point k bad 2 digit lena h
        
plt.show()
```

```python
plt.pie(value,
        labels=key,
       autopct="%0.2f%%",
        startangle=180)
        
plt.show()
```

```python
plt.pie(value,
        labels=key,
        explode=[0.1 for i in range (6)],
        autopct="%0.2f%%",
        shadow=True,
        startangle=180)
plt.show()
        
```
